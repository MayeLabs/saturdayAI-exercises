## **Ejercicio 3**
Crear la lista `subjects` de la primera pregunta y codificar un programa que realice las siguientes acciones:

1. Preguntar al usuario que nota ha obtenido en cada una de las asignaturas.
2. Almacenar las respuestas en una lista llamada `scores`.
3. Imprimir en pantalla la frase `En ASIGNATURA has obtenido un X`, donde `ASIGNATURA` es el nombre de la asignatura y `X` es la nota indicada por el usuario. Mostrar este mensaje para cada una de las asignaturas.

```python
subjects = ['Matemáticas', 'Física', 'Química', 'Historia', 'Lengua']
scores = []
for subject in subjects:
  score = int(input(f'¿Qué nota has obtenido en {subject}? '))
  scores.append(score)

for index, subject in enumerate(subjects):
  print(f'En {subject} has obtenido un {scores[index]}')

# Otra forma - range(len(subjects)
for index in range(len(subjects)):
  print(f'En {subjects[index]} has obtenido un {scores[index]}')
```

```python
# @title Solución Ejercicio 3
subjects = ["Matemáticas", "Física", "Química", "Historia", "Lengua"]
scores = []
for subject in subjects:
    score = input("¿Qué nota has obtenido en " + subject + "? ")
    scores.append(score)
for i in range(len(subjects)):
    print("En " + subjects[i] + " has obtenido un " + scores[i])
```

## **Ejercicio 5**
Escribir un programa que solicite al usuario un número mayor al 1. Luego, debe sumar cada número entero entre el 1 y el número que indicó el usuario (por ejemplo, si el usuario escribió el 4, se debe sumar el 1, 2, 3 y 4, dando como resultado 10). Finalmente, se debe mostrar por pantalla el mensaje `El resultado de la suma sucesiva es X`, donde `X` es el resultado obtenido previamente

```python
num = 1
while num < 2:
  num = int(input("Ingresa un número mayor a 1:"))
sum =  0

for i in range(1, num+1):
  sum = sum + i

print(f'El resultado de la suma sucesiva es {sum}')

```

```python
# @title Solución Ejercicio 5
num = int(input("Ingresa un número mayor a 1: "))
sum = 0
for i in range(1, num + 1):
    sum += i
print("El resultado de la suma sucesiva es", sum)
```
