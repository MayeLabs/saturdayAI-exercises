** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

```python
df['BasePay'].unique()
df['BasePay_float'] = pd.to_numeric(df['BasePay'], errors="coerce")
df.groupby('Year')['BasePay_float'].mean()
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

Mi nota: filter() no acepta una serie booleana directamente.

```python
df.head()
df_year_2013 = df[df['Year'] == 2013]
jobemployee_count = df_year_2013.groupby('JobTitle')['EmployeeName'].count()
#type(jobemployee_count)
#print(jobemployee_count)
count_one = jobemployee_count[jobemployee_count == 1]
print(count_one.count())

```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
chief = df['JobTitle'].str.contains(r'\bChief\b').sum()
print(chief)
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

La pregunta es: **¿los puestos con nombres más largos tienden a tener salarios más altos?**



```python
# Valor de correlacion
df['JobTitle_len'] = df['JobTitle'].str.len()
df.head(3)
print(df['JobTitle_len'].corr(df['TotalPayBenefits']))
```

```python
# Una matriz de correlación
df_corr = df[['JobTitle_len','TotalPayBenefits']]
df_corr.corr()

```

```python
# Forma con apply
df['JobTitle_len'] = df['JobTitle'].apply(lambda x: len(x))
df[['JobTitle_len','TotalPayBenefits']].corr()
```
