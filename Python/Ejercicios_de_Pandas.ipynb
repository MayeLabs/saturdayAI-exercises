## **Ejercicio 4**
Vuelva a crear el mismo DataFrame del ejercicio 1 (Puede copiar y pegar el código de la solución del ejercicio 1). Luego, filtre las filas donde la `"Edad"` sea mayor a 25 y la `"Ciudad"` sea Barcelona. Muestre las filas resultantes por pantalla (Se espera que en cada fila se muestre el nombre, la edad y la ciudad. No es suficiente con mostrar solo la edad).

#### ***Mi solución***

```python
data = {"Nombre":["Ana", "Luis", "Pedro"],
        "Edad": [20, 30, 40],
        "Ciudad": ["Madrid", "Barcelona", "Valencia"] }
df = pd.DataFrame(data)
print(df[(df["Edad"] > 25) & (df["Ciudad"] == "Barcelona")])
```

```python
# @title Solución Ejercicio 4
import pandas as pd

data = {"Nombre": ["Ana", "Luis", "Pedro"], "Edad": [20, 30, 40], "Ciudad": ["Madrid", "Barcelona", "Valencia"]}
df = pd.DataFrame(data)
print(df[(df["Edad"] > 25) & (df["Ciudad"] == "Barcelona")])
```

## **Ejercicio 5**
Vuelva a crear el mismo DataFrame del ejercicio 1 (Puede copiar y pegar el código de la solución del ejercicio 1). Luego, ordene el DataFrame por los valores de la columna `"Edad"` de mayor a menor y muestre el resultado por pantalla.

#### ***Mi solución***

```python
data = {"Nombre":["Ana", "Luis", "Pedro"],
        "Edad": [20, 30, 40],
        "Ciudad": ["Madrid", "Barcelona", "Valencia"] }
df = pd.DataFrame(data)
print(df.sort_values("Edad", ascending=False))
```
