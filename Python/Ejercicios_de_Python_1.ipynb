## **Ejercicio 6**
Escribir un programa que pida al usuario su peso (en kg) y estatura (en metros). Luego, calcule el índice de masa corporal con dichos datos y almacene el resultado en una variable (El índice de masa corporal se calcula dividiendo el peso de una persona para el cuadrado de su altura). Finalmente, muestre por pantalla la frase `Tu índice de masa corporal es X`, donde `X` es el índice de masa corporal calculado y redondeado con dos decimales.

```python
weight = float(input("¿Cuál es tu peso en kg? "))
height = float(input("¿Cuál es tu estatura en metros? "))
indice = round(weight/height**2,2)
print("Tu índice de masa corporal es " + str(indice))
## Otra forma
print("#2 - Tu indice de masa corporal es ", indice)
print(f"#3 - Tu indice de masa corporal es {indice}")
```

```python
# @title Solución Ejercicio 6
weight = input("¿Cuál es tu peso en kg? ")
height = input("¿Cuál es tu estatura en metros? ")
bmi = round(float(weight)/float(height)**2,2)
print("Tu índice de masa corporal es " + str(bmi))
```

## **Ejercicio 7**
Escribe un programa que pida al usuario dos números y guarde ambos números en variables diferentes. Luego, calcule su suma y muestre por pantalla `El resultado de la suma es X`, donde `X` es el resultado de la suma realizada previamente.

```python
a = int(input("Ingresa el primer número: "))
b =  int(input("Ingresa el primer número: "))
x = a + b
print("El resultado de la suma es  ", x)
print(f"El resultado de la suma es  {x}")
print("El resultado de la suma es  " + str(x))
```

```python
# @title Solución Ejercicio 7
num1 = int(input("Ingresa el primer número: "))
num2 = int(input("Ingresa el segundo número: "))
suma = num1 + num2
print("El resultado de la suma es ", suma)
```

## **Ejercicio 8**
Crea un programa que solicite al usuario una cantidad de minutos. Luego, convierta dicha cantidad de minutos en horas y minutos. Finalmente, muestre por pantalla `El tiempo ingresado equivale a X horas y Y minutos`, donde `X` sea la cantidad de horas y `Y` sea la cantidad de minutos luego de la conversión.

```python
time = int(input("Ingresa la cantidad de minutos:"))
print(f"El tiempo ingresado equivale a {time//60} horas y {time%60} minutos")
# Otra forma
horas = time // 60
minutos_restantes = time % 60
print("El tiempo ingresado equivale a", horas, "horas y", minutos_restantes, "minutos")
```

```python
# @title Solución Ejercicio 8
minutos_totales = int(input("Ingresa la cantidad de minutos: "))
horas = minutos_totales // 60
minutos_restantes = minutos_totales % 60
print("El tiempo ingresado equivale a", horas, "horas y", minutos_restantes, "minutos")
```
